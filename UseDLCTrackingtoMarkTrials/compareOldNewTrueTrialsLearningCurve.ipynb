```python
import pandas as pd
import numpy as np
from scipy import io
import os
import glob
import matplotlib.pyplot as plt
```

```python
# direc1 = './dataFolders/DLCtrackingResults/AlternateStep3/

direc1 = '../dataFolders/DLCtrackingResults/AlteranteStep5/'
direc2 = '../dataFolders/Output/Step5_FilesWith_TrueTrialAnd_ProboscisDetect/'

file_old = glob.glob(direc2 + '*.csv')
file_new = glob.glob(direc1 + '*.csv')
```

```python
shapes = ['c-1_', 'c-2_', 'c-3_', 'c-10_']

fig1, ax1 = plt.subplots(2,2, sharex = True, sharey = True, figsize = (20, 10))
ax1 = ax1.ravel()
fig1.suptitle('new')

fig2, ax2 = plt.subplots(2,2, sharex = True, sharey = True, figsize = (20,10))
ax2 = ax2.ravel()
fig2.suptitle('old')

i = 0
for shape in shapes:
    subfile_new = [file for file in file_new if shape in file]
    for file in subfile_new:
        df = pd.read_csv(file)
        
        matched_old = [f for f in file_old if os.path.basename(file) in f]
        df_old = pd.read_csv(matched_old[0])
        
        ax1[i].plot(df.index, df.DiscoveryTime, 'o', label = os.path.basename(file)[:-30])
        ax1[i].legend()
        ax2[i].plot(df_old.index, df_old.DiscoveryTime, 'o', label = os.path.basename(matched_old[0])[:-30])
        ax2[i].legend()
    ax1[i].set_title(shape)
    ax2[i].set_title(shape)
#     plt.legend()
    i+=1
    
fig1.savefig('../dataFolders/DLCtrackingResults/AlteranteStep5/newDiscoveryTime.png')
fig2.savefig('../dataFolders/DLCtrackingResults/AlteranteStep5/oldDiscoveryTime.png')
```

## compare total number of visits at different points in analysis pipeline

```python
df = pd.read_csv(r"../dataFolders/Output/Step3_TrialLabelling_v2/trialSelectionStats.csv")
```

```python
direc = r"../dataFolders/Output/Step5_FilesWith_TrueTrialAnd_ProboscisDetect/"
files = glob.glob(direc + '*.csv')
```

```python
df
```

```python
for name in df.name:
    matched_file = [f for f in files if name in f]
    temp = pd.read_csv(matched_file[0])
    
    length = len(temp)
    df.loc[df.name == name, 'svm'] = length
```

```python
df = df.drop(columns = ['Unnamed: 0'])
df.to_csv(r"../dataFolders/Output/Step3_TrialLabelling_v2/trialSelectionStats.csv")
```

```python
shapes = ['c-1_', 'c-2_', 'c-3_', 'c-10_']

temp = df.copy()
temp.dlc_filter = temp.dlc_filter/temp.initial_num
temp.total_len_filter = temp.total_len_filter/temp.initial_num
temp.final_num = temp.final_num/temp.initial_num
temp['flower'] = temp['name'].map(lambda x: x[0:4])
```

```python
types = ['c-1_', 'c-2_', 'c-3_', 'c-10']

f1, ax = plt.subplots(4,1, figsize = (8,6*4), sharey = True)
ax = ax.ravel()
i = 0
for types in types:
    test = temp[temp['flower'] == types]
    for _,row in test.iterrows():
        ax[i].plot(['proboscis', 'dwell time', 'merge'], np.asarray([row.dlc_filter, row.total_len_filter, row.final_num])*100.0, 'o-')
    ax[i].set_title(types)
    i+=1
```
